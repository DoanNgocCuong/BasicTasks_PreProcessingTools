- **GPT-3.5-Turbo**: $0.0015 per 1,000 tokens for prompt input and $0.002 per 1,000 tokens for completion.

In comparison, generating a single image with DALL-E 3 at standard quality can be more costly than processing several thousand tokens with GPT-4 or GPT-3.5-Turbo.

````python
# Test: context = "Minh and Nhi are having a first date"

import os
import requests
import json
from openai import OpenAI

# Initialize the OpenAI client
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)

def get_completion(prompt, model="gpt-4o", temperature=0):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content


def create_prompt_Dalle3_from_context(context): 
    """
    context for gen best_prompt for DALL-E-3
    """

    # Create the prompt for creating a fill-in-the-blank exercise
    prompt_Dalle3_from_context = f"""

    # Role: You are an expert in writing prompts for DALL-E-3, and your prompts result in beautiful and accurate images.
    # Context: Given a specific context {context}, generate a detailed and vivid prompt for DALL-E-3 that will be used to create an image
    Ensure the prompt is clear, descriptive, and captures the essence of the given context.

    # Instruction: 
    # Step 1: Conceptualize the Image:
        - Based on {context}, imagine you are an artist creating a visual representation of this context.
        - Focus on ensuring the idea reflects {context} accurately.

    # Step 2: Detailed Prompt Creation
    Write prompt to create an image for the context {context} with the following formula:
    1. Type: Digital Minimalist Illustration - Clean Lines and Shapes - Flat Design - Soft Shading.
    2. Subject: Từ step 1 => 'Conceptualize the Image'.
    3. Features: Focused on {context} with the following elements:
    - Characters: Facial expressions and positions that perfectly match {context}.
    4. Setting: Setting: A plain background with no patterns, only include two small elements surrounding the characters. 
    5. Mood/Color: Use only 3-4 pastel colors, create a warm and friendly atmosphere with a well-balanced contrast to highlight {context}.
    6. Style: Keep the style clean and simple.
    7. Emphasis: Highlight context {context} as the image's main focus. 
    8. Note: 
        - Do not include any words or text in the image. 
        - Nhân vật gương mặt xinh xắn hơn, hiền lành, vui tươi hơn, quần áo kín đáo không nhạy cảm phản cảm


    Input: "{context}"
    Output: Prompt 

    ### Phong cách ảnh follow theo các ảnh đã tạo sau: 
0. A minimalist digital illustration of Minh and Nhi having their first date in a simple outdoor setting. They are sitting at a small café table with drinks in front of them. The background is plain with very few details, using no more than 3-4 pastel colors. Minh is wearing a casual shirt and jeans, while Nhi is wearing a beautiful dress. Both are smiling and engaged in a pleasant conversation. The style is clean and simple.

1. A digital illustration in a modern cartoon style featuring two new friends standing outside a door with a 'WELCOME' sign. The woman has dark skin, short curly hair, wearing a white tank top, beige high-waisted pants, and gold hoop earrings. The man has light skin, black hair, a beard, wearing a brown shirt over a white t-shirt, and blue jeans. Both are smiling warmly and look friendly. The background includes some potted plants and a light-colored wall.

2. A digital illustration in a modern cartoon style featuring an Asian woman, the new staff member in your company, sitting at a desk in an office setting. She has light skin, short black hair, wearing a high-neck white blouse and beige high-waisted pants. She is smiling warmly and looks friendly. The background includes a desk with a computer and some potted plants, creating a typical office environment.

3. A digital illustration in a modern cartoon style featuring an Asian man in a Pho restaurant. He has light skin, short black hair, and is wearing a high-neck white shirt and beige pants. He is smiling warmly and looks friendly while eating a bowl of Pho. The background includes a table with a bowl of Pho, some potted plants, and a small Vietnamese flag with a red background and yellow star, creating a typical Pho restaurant environment.

4. A minimalist digital illustration of a young woman deeply in love with playing the guitar, set in a simple, clean background. The woman has light skin, long black hair, and is wearing a pastel-colored dress. She is sitting cross-legged on the floor, gently strumming an acoustic guitar with a serene and content expression on her face. Surrounding her are two small elements: a potted plant and a small stack of music sheets. The background is plain with no patterns, using only 3-4 pastel colors to create a warm and friendly atmosphere. The style is clean and simple, with soft shading and flat design, emphasizing the woman's connection to the guitar and music as the main focus of the image. No words or text are included.

5. Digital minimalist illustration with clean lines and shapes, flat design, and soft shading. The subject is a woman looking stressed at her work desk, surrounded by documents and a laptop. Her facial expression shows stress, with a hand on her forehead. Beside her, an image of the same woman relaxed on a sofa, holding a cup of tea, smiling, and looking content. The background is plain with no patterns, featuring only two small elements: a clock on the wall and a plant in a pot. The color palette consists of 3-4 pastel colors, creating a warm and friendly atmosphere with well-balanced contrast. The style is clean and simple, highlighting the concept of work-life balance and relaxation after work. The characters have attractive, kind, and cheerful faces, with modest, non-revealing clothing. Make the design flatter and smoother.
    
    # OUTPUT ONLY: chỉ trả về JSON format: {{"prompt": "best prompt of image"}}
    """

    output = get_completion(prompt_Dalle3_from_context, temperature=0)

    # Extract and parse the JSON string
    response_json = json.loads(output.strip().strip('```json').strip().strip('```'))
    best_prompt = response_json["prompt"]
    print(f"The extracted prompt is: {best_prompt}")
    return best_prompt

def generate_and_save_image(prompt, image_name, STT_row=None, image_index=0, output_folder=None):
    """
    prompt: The generated prompt for the image
    image_name: The context to be used for naming the image
    STT_row: The row number of the STT file excel sheet
    image_index: The index of the image to be saved, image1, 2, 3
    output_folder: The folder where images will be saved
    """
    response = client.images.generate(
        model="dall-e-3", 
        prompt=prompt,
        n=1,  # Number of images to generate (only 1 supported)
        size="1024x1024"
    )
    image_url = response.data[0].url
    print(f"Generated Image URL {image_index + 1}: {image_url}")
    image_response = requests.get(image_url)
    if image_response.status_code == 200:
        context_words = '_'.join(image_name.split()[:7]).lower()
        image_filename = f'{STT_row}_{context_words}_{image_index + 1}.png'
        image_path = os.path.join(output_folder, image_filename)
        with open(image_path, 'wb') as file:
            file.write(image_response.content)
        print(f"Image saved successfully as '{image_path}'")
    else:
        print(f"Failed to retrieve the image {image_index + 1}")
    return image_filename


context = " Hình ảnh làm nổi bật: He lives in a house. He doesn't want to live in an apartment."
# Generate the prompt
best_prompt = create_prompt_Dalle3_from_context(context)
# Create a folder to store the images
output_folder = 'context_images'
os.makedirs(output_folder, exist_ok=True)

# Generate and save the image

for i in range(3): 
    generate_and_save_image(prompt=best_prompt, image_name=context, output_folder=output_folder, image_index=i)
````

````python
# Test: context = "Minh and Nhi are having a first date"

import os
import requests
import json
from openai import OpenAI

# Initialize the OpenAI client
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)

def get_completion(prompt, model="gpt-4o", temperature=0):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content


def create_prompt_Dalle3_for_avatar_image(context): 
    """
    context for gen best_prompt for DALL-E-3
    """

    # Create the prompt for creating a fill-in-the-blank exercise
    prompt_Dalle3_for_image = f"""

    # Role: You are an expert in writing prompts for DALL-E-3, and your prompts result in beautiful and accurate avatar images.
    # Context: Given a specific context {context}, generate a detailed and vivid prompt for DALL-E-3 that will be used to create an avatar image
    Ensure the prompt is clear, descriptive, and captures the essence of the given context.

# Instruction:
# Step 1: Conceptualize the Image:
    - Based on {context}, imagine you are an artist creating a visual representation of this context.
    - Focus on ensuring the idea reflects {context} accurately.

# Step 2: Detailed Prompt Creation
Write prompt to create an avatar image for the context {context} with the following formula:
1. Type: Digital Minimalist Illustration - Clean Lines and Shapes - Flat Design - Soft Shading.
2. Subject: Avatar Image
3. Features: Characters: Facial expressions and positions that perfectly match {context}.
4. Setting: No setting for Avatar Image
5. Mood/Color: Use only 3-4 pastel colors, create a warm and friendly atmosphere with a well-balanced contrast to highlight {context}.
6. Style: Keep the style clean and simple.
7. Emphasis: Avatar Image



    Input: "{context}"
    Output: Prompt 

    ### Phong cách ảnh follow theo các ảnh đã tạo sau: 
Digital Minimalist Illustration - Clean Lines and Shapes - Flat Design - Soft Shading. Create an avatar image of a female close friend. The character should have a warm and friendly facial expression, with a gentle smile and bright eyes. Use only 3-4 pastel colors, such as soft pink, light blue, and pale yellow, to create a warm and friendly atmosphere. Ensure the colors are well-balanced to highlight the character. Keep the style clean and simple, with no background setting. Emphasize the avatar image.

    # OUTPUT ONLY: chỉ trả về JSON format: {{"prompt": "best prompt of image"}}
    """

    output = get_completion(prompt_Dalle3_for_image, temperature=0)

    # Extract and parse the JSON string
    response_json = json.loads(output.strip().strip('```json').strip().strip('```'))
    best_prompt = response_json["prompt"]
    print(f"The extracted prompt is: {best_prompt}")
    return best_prompt

def generate_and_save_image(prompt, image_name, STT_row=None, image_index=0, output_folder=None):
    """
    prompt: The generated prompt for the image
    image_name: The context to be used for naming the image
    STT_row: The row number of the STT file excel sheet
    image_index: The index of the image to be saved, image1, 2, 3
    output_folder: The folder where images will be saved
    """
    response = client.images.generate(
        model="dall-e-3", 
        prompt=prompt,
        n=1,  # Number of images to generate (only 1 supported)
        size="1024x1024"
    )
    image_url = response.data[0].url
    print(f"Generated Image URL {image_index + 1}: {image_url}")
    image_response = requests.get(image_url)
    if image_response.status_code == 200:
        context_words = '_'.join(image_name.split()[:7]).lower()
        image_filename = f'{STT_row}_{context_words}_{image_index + 1}.png'
        image_path = os.path.join(output_folder, image_filename)
        with open(image_path, 'wb') as file:
            file.write(image_response.content)
        print(f"Image saved successfully as '{image_path}'")
    else:
        print(f"Failed to retrieve the image {image_index + 1}")
    return image_filename


context = "A female close friend"
# Generate the prompt
best_prompt = create_prompt_Dalle3_for_avatar_image(context)
# Create a folder to store the images
output_folder = 'avatar_images'
os.makedirs(output_folder, exist_ok=True)

# Generate and save the image

for i in range(3): 
    generate_and_save_image(prompt=best_prompt, image_name=context, output_folder=output_folder, image_index=i)
````
