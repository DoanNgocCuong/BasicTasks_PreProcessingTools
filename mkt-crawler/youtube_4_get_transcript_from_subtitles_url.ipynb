```python
import json
import requests
import xml.etree.ElementTree as ET
from datetime import timedelta
from airtable import Airtable
import time

# Airtable configuration
AIRTABLE_BASE_ID = 'app7mcJMf9O8GQRal'
AIRTABLE_API_KEY = ''
TABLE_NAME = 'youtube_post'
ata = Airtable(AIRTABLE_BASE_ID, TABLE_NAME, AIRTABLE_API_KEY)

# Lấy danh sách video_id từ Airtable
records = ata.get_all(view='1_sub')
video_ids = [record['fields'].get('video_id') for record in records if record['fields'].get('video_id')]

for video_id in video_ids:
    retries = 3
    for attempt in range(retries):
        # Tìm record tương ứng với video_id
        formula = f"{{video_id}} = '{video_id}'"
        matching_records = ata.get_all(formula=formula)
        
        if matching_records and 'subtitles_url' in matching_records[0]['fields']:
            subtitle_url = matching_records[0]['fields']['subtitles_url']
            transcript_lines = []
            
            # Lấy và xử lý XML từ subtitles_url
            subtitle_response = requests.get(subtitle_url)
            if subtitle_response.status_code == 200:
                root = ET.fromstring(subtitle_response.content)
                for elem in root.iter('text'):
                    start = elem.get('start')
                    text = elem.text
                    if start and text:
                        start_seconds = float(start)
                        start_time = str(timedelta(seconds=start_seconds))
                        hh_mm_ss = start_time.split('.')[0]
                        transcript_lines.append(f"[{hh_mm_ss}] {text}")
                
                if transcript_lines:
                    transcript_text = '\n'.join(transcript_lines)
                    
                    # Update Airtable record với script đầy đủ
                    update_fields = {
                        'script': transcript_text,
                        'script_status': 'Completed'
                    }
                    
                    record_id = matching_records[0]['id']
                    ata.update(record_id, update_fields)
                    print(f"Updated transcript for video ID {video_id}")
                break
            else:
                print(f"Failed to retrieve subtitle data from URL: {subtitle_url}")
                if attempt < retries - 1:
                    time.sleep(2)
                else:
                    print(f"All retry attempts failed for video ID: {video_id}")
                    record_id = matching_records[0]['id']
                    ata.update(record_id, {'script_status': 'Failed to retrieve transcript'})
        else:
            print(f"No subtitles_url found for video ID: {video_id}")
            if matching_records:
                record_id = matching_records[0]['id']
                ata.update(record_id, {'script_status': 'No subtitles URL available'})
            break
```
